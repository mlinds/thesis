```python
import xarray as xr
import numpy as np
from atl_module import load_netcdf, icesat_bathymetry
```

```python
beamdata = load_netcdf.load_beam_array_ncds(
    "../data/test_sites/florida_keys/ATL03/processed_ATL03_20181017085235_02860107_005_01.nc",
    "gt2l",
)
```

```python
df = icesat_bathymetry.add_along_track_dist(beamdata).set_index("dist_or")
```

```python
df.Z_geoid.index
```

```python
def printfunc(series_in):
    center_loc = series_in.index[int(len(series_in) / 2)]
    # print(center_loc)
    window_distance = 100
    min_periods = 50
    series_in = series_in[
        (series_in.index > center_loc - window_distance / 2)
        & (series_in.index < center_loc + window_distance / 2)
    ]
    # print(series_in)
    if len(series_in) >= min_periods:
        return series_in.count()
    else:
        return np.NaN


out = df.Z_geoid.rolling(
    window=201,
    center=True,
).apply(printfunc)
```
