```python
session = requests.session()
capability_url = f"https://n5eil02u.ecs.nsidc.org/egi/capabilities/ATL03.005.xml"
s = session.get(capability_url)
response = session.get(s.url, auth=(EARTHDATA_USERNAME, EARTHDATA_PASSWORD))
```

```python
def parse_capabilities_xml(xmlstring):
    testtree = etree.fromstring(response.content)
    projlist = [x.attrib['value'] for x in testtree.iterdescendants(tag='Projection')]
    formatlist = [x.attrib['value'] for x in testtree.iterdescendants(tag='Format')]
    varlist = [x.attrib['value'].replace(':','/') for x in testtree.iterdescendants(tag='SubsetVariable')]
    return projlist,formatlist,varlist
    
```

```python
projections, formats, variables = parse_capabilities_xml(response.content)
```
