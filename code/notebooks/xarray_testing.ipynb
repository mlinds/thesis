```python
import numpy as np
from netCDF4 import Dataset
%cd ..
```

```python
ncfile = '../data/test_sites/florida_keys/ATL09/processed_ATL09_20181017081233_02860101_005_01.nc'
ds = Dataset(ncfile)
groups = ds.groups
grouplist = [profname for profname in groups.keys() if profname.startswith('prof')]
```

```python
xrlist = []
for groupname in grouplist:
    for subgroup in groups[groupname].groups.keys():
        xrds = Dataset(ncfile).groups[groupname].groups[subgroup]
        xrlist.append(xrds)
```

```python
for group in xrlist:
    for variable in group.variables.keys():
        name=group.variables[variable].name
        type_=group.variables[variable].dtype
        array = group.variables[variable][:]
    columndtype = np.dtype([(name,type_)])
    
    columnstarray = np.array(array,dtype=columndtype)
    print(columnstarray.shape)
```
